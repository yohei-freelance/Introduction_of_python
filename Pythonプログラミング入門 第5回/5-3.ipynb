# json ファイルのダンプとロード
json は、各種のデータを記録するためのフォーマットです。特に、辞書や辞書のリストを記録することができます。

json モジュールを用いることにより、Pythonの各種のデータをファイルに書き出す（ダンプする）ことができ、また、ファイルからロードすることができます。

```python
import json
```

```python
d = {'apple' : 3, 'pen' : 5}
```

```python
with open("test.json", "w") as f:
    json.dump(d, f)
```

```python
with open("test.json", "r") as f:
    d1 = json.load(f)
```

```python
d1
```

```python
d == d1
```

# 練習

1. 以下のリスト内包の結果を fib.json というファイルに json フォーマットでダンプしてください。

2. ダンプしたファイルからロードして、同じものが得られることを確かめてください。

```python
def fib(n):
    if (n == 0):
        return 0
    elif (n == 1):
        return 1
    else:
        return fib(n-1)+fib(n-2)

d = [{'n': n, 'fib' : fib(n)} for n in range(0,10)]

with open("fib.json", "w") as f:
    json.dump(d,f)
    
with open("fib.json", "r") as f:
    d1 = json.load(f)
    
d == d1
```

```python
#予習課題5-2

search_catalog("Python")
```

```python
with open("python_courses.json", "w", encoding="utf-8") as f:
    json.dump(search_catalog("Python"))
```


# [参考] 不要な空白や改行
ファイルから読み込んだ文字列の前後に不要な空白や改行がある場合は、組み込み関数strip()を使用するとそれらの空白・改行を除去することができます。

```python
"   This is strip.\n".strip()
```

lstrip は文頭、rstrip は文末の空白や改行を除去します。

```python
"   This is strip.\n".lstrip()
```

```python
"   This is strip.\n".rstrip()
```
